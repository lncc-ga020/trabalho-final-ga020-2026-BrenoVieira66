# Equações de Euler 1D com termos fonte

O ponto de partida do modelo é o sistema unidimensional de conservação de massa, quantidade de movimento e energia total. Para uma tubulação, escrevemos:

$$
\frac{\partial \rho}{\partial t}
+
\frac{\partial(\rho u)}{\partial x}
= 0,
$$

$$
\frac{\partial(\rho u)}{\partial t}
+
\frac{\partial(\rho u^2 + p)}{\partial x}
= \rho g_x - \mathcal{F},
$$

$$
\frac{\partial E}{\partial t}
+
\frac{\partial((E+p)u)}{\partial x}
= Q.
$$

Neste sistema:

- $\rho$ é a massa específica da substância no fluido;
- $u$ é a velocidade do fluido;
- $p$ é a pressão;
- $E$ é a energia total;
- $g_x$ é a componente da gravidade na direção do tubo;
- $\mathcal{F}$ representa a perda de quantidade de movimento por atrito;
- $Q$ representa um termo fonte de energia, por exemplo troca de calor ou dissipação.

A primeira equação conserva massa. A segunda descreve o balanço de quantidade de movimento, incluindo forças de campo e atrito. A terceira descreve o balanço de energia total.

## Teorema de Transporte de Reynolds

Antes de escrever as equações diferenciais de conservação, é útil lembrar o Teorema de Transporte de Reynolds. Ele conecta duas formas de descrever o movimento de um fluido:

- acompanhar um sistema material, isto é, sempre as mesmas partículas de fluido;
- observar um volume de controle fixo no espaço, como um trecho da tubulação.

Para uma propriedade extensiva $B$, associada a uma propriedade por unidade de massa $b$, podemos escrever:

$$
B = \int_{V(t)} \rho b \, dV.
$$

Uma forma usual do Teorema de Transporte de Reynolds é:

$$
\frac{dB_{\text{sistema}}}{dt}
=
\frac{\partial}{\partial t}
\int_{VC} \rho b \, dV
+
\int_{SC} \rho b(\mathbf{u}\cdot\mathbf{n})\,dA.
$$

Aqui, $VC$ é o volume de controle, $SC$ é a superfície de controle, $\mathbf{u}$ é a velocidade do fluido e $\mathbf{n}$ é o vetor normal externo à superfície.

A interpretação é: a variação de uma propriedade no sistema material pode ser escrita como a variação dentro do volume de controle mais o fluxo líquido dessa propriedade pelas fronteiras.

No caso unidimensional de uma tubulação, essa ideia se reduz a um balanço em um intervalo $[x_1,x_2]$:

$$
\frac{d}{dt}
\int_{x_1}^{x_2}
(\text{quantidade conservada})\,dx
=
\text{fluxo que entra}
-
\text{fluxo que sai}
+
\text{fontes internas}.
$$

Essa é a base física das formas integrais usadas nas deduções de massa, quantidade de movimento e energia. A partir dessas formas integrais, obtemos as equações diferenciais conservativas do modelo de Euler 1D.


## Interpretação e dedução das equações

Seguindo a abordagem apresentada por LeVeque em *Finite Volume Methods for Hyperbolic Problems*, o sistema acima pode ser visto como uma lei de conservação hiperbólica com termos fonte. A estrutura geral é:

$$
\frac{\partial U}{\partial t}
+
\frac{\partial F(U)}{\partial x}
= S(U,x,t),
$$

onde $U$ representa as variáveis conservadas, $F(U)$ representa os fluxos que atravessam as fronteiras de cada volume de controle, e $S(U,x,t)$ reúne os efeitos que não aparecem como fluxo convectivo, como gravidade, atrito e troca de calor.

### Conservação de massa

Uma forma de entender essa equação é começar pela massa total contida em um intervalo $[x_1,x_2]$. Como $\rho(x,t)$ representa a massa específica do fluido, a massa total nesse trecho da tubulação é:

$$
\int_{x_1}^{x_2} \rho(x,t)\,dx.
$$

Essa massa só muda por causa do fluxo de massa que entra ou sai pelas extremidades do intervalo. O fluxo de massa em uma posição $x$ é dado por:

$$
\text{fluxo de massa} = \rho(x,t)u(x,t).
$$

Assim, o balanço de massa no intervalo pode ser escrito como:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} \rho(x,t)\,dx
=
\rho(x_1,t)u(x_1,t)
-
\rho(x_2,t)u(x_2,t).
$$

Usando a notação de cálculo para escrever a diferença entre os fluxos nas extremidades, temos:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} \rho(x,t)\,dx
=
-
\int_{x_1}^{x_2}
\frac{\partial}{\partial x}(\rho u)\,dx.
$$

Reunindo os termos em uma única integral:

$$
\int_{x_1}^{x_2}
\left[
\frac{\partial \rho}{\partial t}
+
\frac{\partial(\rho u)}{\partial x}
\right]dx
=0.
$$

Como isso deve valer para qualquer intervalo $[x_1,x_2]$, o integrando deve ser nulo. Portanto, obtemos a forma diferencial da conservação de massa:

$$
\frac{\partial \rho}{\partial t}
+
\frac{\partial(\rho u)}{\partial x}
=0.
$$

Essa equação afirma que a massa dentro de um volume de controle só muda por causa do fluxo de massa que entra ou sai pelas suas fronteiras. No método de volumes finitos, essa é exatamente a ideia central: atualizar a média da célula a partir da diferença entre o fluxo na face esquerda e o fluxo na face direita.

### Conservação da quantidade de movimento

A dedução segue a mesma lógica da conservação de massa, mas agora a grandeza conservada é a quantidade de movimento por unidade de volume, $\rho u$. Portanto, a quantidade de movimento total no intervalo $[x_1,x_2]$ é:

$$
\int_{x_1}^{x_2} \rho(x,t)u(x,t)\,dx.
$$

Essa quantidade muda devido aos fluxos de quantidade de movimento nas extremidades e também devido às forças externas que atuam no volume de controle. O fluxo de quantidade de movimento é:

$$
\text{fluxo de quantidade de movimento} = \rho u^2 + p.
$$

Assim, o balanço integral pode ser escrito como:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} \rho u\,dx
=
\left(\rho u^2+p\right)(x_1,t)
-
\left(\rho u^2+p\right)(x_2,t)
+
\int_{x_1}^{x_2}\left(\rho g_x-\mathcal{F}\right)dx.
$$

Usando a derivada espacial para representar a diferença dos fluxos nas extremidades:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} \rho u\,dx
=
-
\int_{x_1}^{x_2}
\frac{\partial}{\partial x}\left(\rho u^2+p\right)dx
+
\int_{x_1}^{x_2}\left(\rho g_x-\mathcal{F}\right)dx.
$$

Reunindo os termos em uma única integral:

$$
\int_{x_1}^{x_2}
\left[
\frac{\partial(\rho u)}{\partial t}
+
\frac{\partial(\rho u^2+p)}{\partial x}
-
\left(\rho g_x-\mathcal{F}\right)
\right]dx
=0.
$$

Como o intervalo é arbitrário, obtemos a forma diferencial:

$$
\frac{\partial(\rho u)}{\partial t}
+
\frac{\partial(\rho u^2+p)}{\partial x}
=
\rho g_x-\mathcal{F}.
$$

Aqui, $\rho u$ é a quantidade de movimento por unidade de volume. O fluxo $\rho u^2 + p$ tem duas contribuições: transporte convectivo de momento e força de pressão. Os termos do lado direito representam fontes externas: $\rho g_x$ acelera o fluido na direção do tubo, enquanto $\mathcal{F}$ representa a perda por atrito com a parede.

Na linguagem de LeVeque, a parte à esquerda é o problema hiperbólico conservativo, associado à propagação de ondas. Já o lado direito é tratado como termo fonte, que pode ser incorporado depois do passo de fluxo por técnicas como splitting.

### Conservação de energia

A dedução novamente segue a ideia de balanço em um intervalo. Agora a grandeza conservada é a energia total por unidade de volume, $E$. Assim, a energia total contida no intervalo $[x_1,x_2]$ é:

$$
\int_{x_1}^{x_2} E(x,t)\,dx.
$$

Essa energia muda devido ao fluxo de energia pelas extremidades e também devido a possíveis fontes internas. O fluxo de energia é:

$$
\text{fluxo de energia} = (E+p)u.
$$

O termo $E u$ representa o transporte de energia pelo movimento do fluido. O termo $p u$ representa o trabalho associado à pressão.

Assim, o balanço integral da energia pode ser escrito como:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} E\,dx
=
\left((E+p)u\right)(x_1,t)
-
\left((E+p)u\right)(x_2,t)
+
\int_{x_1}^{x_2} Q\,dx.
$$

Usando a derivada espacial para representar a diferença entre os fluxos nas extremidades:

$$
\frac{d}{dt}
\int_{x_1}^{x_2} E\,dx
=
-
\int_{x_1}^{x_2}
\frac{\partial}{\partial x}\left((E+p)u\right)dx
+
\int_{x_1}^{x_2}Q\,dx.
$$

Reunindo os termos em uma única integral:

$$
\int_{x_1}^{x_2}
\left[
\frac{\partial E}{\partial t}
+
\frac{\partial((E+p)u)}{\partial x}
-
Q
\right]dx
=0.
$$

Como o intervalo é arbitrário, obtemos a forma diferencial:

$$
\frac{\partial E}{\partial t}
+
\frac{\partial((E+p)u)}{\partial x}
=Q.
$$

A variável $E$ representa a energia total por unidade de volume, somando energia interna e energia cinética. O fluxo $(E+p)u$ representa o transporte de energia pelo movimento do fluido e o trabalho associado à pressão. O termo $Q$ pode representar troca de calor, dissipação ou outro mecanismo de adição ou remoção de energia.

Essa forma é adequada para métodos conservativos porque choques, rarefações e descontinuidades de contato podem aparecer naturalmente na solução. Por isso, em vez de discretizar diretamente derivadas ponto a ponto, o método de volumes finitos calcula fluxos numéricos nas interfaces entre células, geralmente resolvendo ou aproximando problemas de Riemann locais.

## Hipóteses iniciais de simplificação

Nesta primeira formulação, iremos considerar que a tubulação é totalmente horizontal. Portanto, a componente da gravidade na direção axial é nula:

$$
g_x = g\sin(\theta) = g\sin(0) = 0.
$$

Também será adotada, inicialmente, a hipótese de escoamento adiabático, isto é, sem troca de calor com o ambiente. Assim:

$$
Q = 0.
$$

Diferentemente da etapa exploratória com advecção escalar, no sistema HEM inspirado na referência [5] não impomos velocidade constante. A velocidade passa a ser calculada a partir das variáveis conservadas:

$$
u = \frac{m}{\rho},
\qquad
m=\rho u.
$$

Com essas hipóteses, o sistema fica:

$$
\frac{\partial \rho}{\partial t}
+
\frac{\partial(\rho u)}{\partial x}
= 0,
$$

$$
\frac{\partial(\rho u)}{\partial t}
+
\frac{\partial(\rho u^2 + p)}{\partial x}
= -\mathcal{F},
$$

$$
\frac{\partial E}{\partial t}
+
\frac{\partial((E+p)u)}{\partial x}
= 0.
$$

Ou seja, nesta etapa mantemos o atrito na equação de quantidade de movimento, mas desconsideramos a gravidade axial e a troca de calor.


# Diferenças Finitas: parte teórica

Nesta etapa, o objetivo é transformar as equações diferenciais parciais em equações algébricas que possam ser calculadas em uma malha. A ideia central das Diferenças Finitas é aproximar derivadas por diferenças entre valores da solução em pontos discretos.

Para uma variável genérica $\phi(x,t)$, escrevemos:

$$
\phi_i^n \approx \phi(x_i,t^n),
$$

onde o índice $i$ indica a posição no espaço e o índice $n$ indica o instante de tempo.

## Método de Euler explícito

O método de Euler explícito aproxima a derivada temporal usando o valor atual da solução:

$$
\frac{\partial \phi}{\partial t}(x_i,t^n)
\approx
\frac{\phi_i^{n+1}-\phi_i^n}{\Delta t}.
$$

A palavra explícito significa que o valor futuro $\phi_i^{n+1}$ é calculado diretamente a partir de valores já conhecidos no tempo $t^n$. Para uma equação do tipo:

$$
\frac{\partial \phi}{\partial t}=R(\phi),
$$

a atualização fica:

$$
\phi_i^{n+1}=\phi_i^n+\Delta t\,R_i^n.
$$

Essa é a forma que usaremos para avançar as equações no tempo.

## Consistência

Um esquema é consistente quando, ao reduzir $\Delta t$ e $\Delta x$, a equação discreta se aproxima da equação diferencial original. Em outras palavras, se substituirmos a solução exata da EDP na fórmula numérica, o erro residual deve tender a zero quando a malha é refinada.

Para Euler explícito, usando expansão de Taylor:

$$
\phi(x_i,t^{n+1})
=
\phi(x_i,t^n+\Delta t).
$$

Expandindo em torno de $t^n$, temos:

$$
\phi(x_i,t^n+\Delta t)
=
\phi(x_i,t^n)
+
\Delta t\frac{\partial \phi}{\partial t}(x_i,t^n)
+
\frac{\Delta t^2}{2}
\frac{\partial^2 \phi}{\partial t^2}(x_i,t^n)
+
O(\Delta t^3).
$$

Subtraindo $\phi(x_i,t^n)$ dos dois lados:

$$
\phi(x_i,t^{n+1})-\phi(x_i,t^n)
=
\Delta t
\frac{\partial \phi}{\partial t}(x_i,t^n)
+
\frac{\Delta t^2}{2}
\frac{\partial^2 \phi}{\partial t^2}(x_i,t^n)
+
O(\Delta t^3).
$$

Dividindo por $\Delta t$:

$$
\frac{\phi_i^{n+1}-\phi_i^n}{\Delta t}
=
\frac{\partial \phi}{\partial t}(x_i,t^n)
+
\frac{\Delta t}{2}
\frac{\partial^2 \phi}{\partial t^2}(x_i,t^n)
+
O(\Delta t^2).
$$

O termo dominante do erro é proporcional a $\Delta t$. Assim, escrevemos:

$$
\frac{\phi_i^{n+1}-\phi_i^n}{\Delta t}
=
\frac{\partial \phi}{\partial t}(x_i,t^n)
+O(\Delta t).
$$

Portanto, Euler explícito é consistente de primeira ordem no tempo.

## Estabilidade

A estabilidade está relacionada ao crescimento dos erros numéricos. Mesmo que um método seja consistente, pequenos erros de arredondamento ou pequenas perturbações podem crescer durante a simulação.

Para problemas de transporte, uma condição fundamental é o número CFL:

$$
\mathrm{CFL}=\frac{|u_0|\Delta t}{\Delta x}.
$$

Esse número mede quantas células a informação percorre em um passo de tempo. Se o passo de tempo for grande demais, a solução numérica tenta avançar mais rápido do que a informação física, e o método pode se tornar instável.

Como regra inicial para advecção explícita, usamos:

$$
\mathrm{CFL}\leq 1.
$$

Uma forma simples de visualizar a estabilidade do Euler explícito é usar o problema modelo:

$$
\frac{d\phi}{dt}=\lambda\phi.
$$

Aplicando Euler explícito:

$$
\frac{\phi^{n+1}-\phi^n}{\Delta t}
=
\lambda\phi^n.
$$

Isolando o valor no próximo passo de tempo:

$$
\phi^{n+1}
=
(1+\lambda\Delta t)\phi^n.
$$

O termo

$$
G=1+\lambda\Delta t
$$

é chamado fator de amplificação. Após vários passos:

$$
\phi^n=G^n\phi^0.
$$

Para que erros ou perturbações não cresçam indefinidamente, é necessário que:

$$
|G|\leq 1.
$$

Portanto, a condição de estabilidade para esse problema modelo é:

$$
|1+\lambda\Delta t|\leq 1.
$$

Quando $\lambda<0$ é real, essa condição leva a:

$$
\Delta t \leq \frac{2}{|\lambda|}.
$$

Assim, o método de Euler explícito não é incondicionalmente estável. Ele é estável apenas quando o passo de tempo respeita uma restrição adequada. Em problemas de transporte, essa restrição aparece na forma da condição CFL.

## Convergência

Convergência significa que a solução numérica se aproxima da solução exata quando refinamos a malha:

$$
\Delta x \to 0,
\qquad
\Delta t \to 0.
$$

A relação entre consistência, estabilidade e convergência é formalizada pelo Teorema de Equivalência de Lax-Richtmyer. Para problemas lineares bem postos, esse teorema afirma que um esquema de diferenças finitas consistente é convergente se, e somente se, for estável. Em termos práticos, consistência garante que a equação discreta aproxima a equação diferencial, enquanto estabilidade impede que os erros cresçam sem controle durante a evolução.

Para o método de Euler aplicado a problemas de evolução, o erro global é de primeira ordem no tempo, isto é:

$$
|e^n| = O(\Delta t).
$$

No caso de EDPs, também precisamos considerar o erro da aproximação espacial.

## Diferença centrada

A diferença centrada usa os pontos dos dois lados de $x_i$:

$$
\frac{\partial \phi}{\partial x}(x_i,t^n)
\approx
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2\Delta x}.
$$

Ela é uma aproximação de segunda ordem no espaço:

$$
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2\Delta x}
=
\frac{\partial \phi}{\partial x}(x_i,t^n)
+O(\Delta x^2).
$$

Apesar disso, para problemas puramente advectivos ou hiperbólicos, a diferença centrada pode gerar oscilações numéricas, especialmente perto de descontinuidades ou gradientes fortes.

## Diferença upwind

A diferença upwind leva em conta a direção do escoamento. Se $u_0>0$, a informação vem da esquerda para a direita, então usamos:

$$
\frac{\partial \phi}{\partial x}(x_i,t^n)
\approx
\frac{\phi_i^n-\phi_{i-1}^n}{\Delta x}.
$$

Se $u_0<0$, a informação vem da direita para a esquerda, então usamos:

$$
\frac{\partial \phi}{\partial x}(x_i,t^n)
\approx
\frac{\phi_{i+1}^n-\phi_i^n}{\Delta x}.
$$

O método upwind é de primeira ordem no espaço, mas costuma ser mais estável para problemas de transporte. Ele introduz difusão numérica, mas reduz oscilações não físicas.

Assim, nesta primeira etapa, a diferença centrada é útil para entender a construção matemática do método, enquanto o upwind é mais adequado quando buscamos uma implementação estável para advecção e ondas propagantes.


# Implementação de Diferenças Finitas

Agora aproximamos o sistema de Euler 1D na forma usada na referência [5] para o modelo HEM. Nesta formulação, a velocidade não é mais prescrita como constante. Em vez disso, resolvemos a quantidade de movimento:

$$
m = \rho u,
$$

logo:

$$
u = \frac{m}{\rho}.
$$

Com tubo horizontal e sem troca de calor, isto é, $g_x=0$ e $Q=0$, o sistema fica:

$$
\frac{\partial \rho}{\partial t}
+
\frac{\partial m}{\partial x}
=0,
$$

$$
\frac{\partial m}{\partial t}
+
\frac{\partial}{\partial x}
\left(\frac{m^2}{\rho}+p\right)
=-\mathcal{F},
$$

$$
\frac{\partial E}{\partial t}
+
\frac{\partial}{\partial x}
\left((E+p)\frac{m}{\rho}\right)
=0.
$$

Os fluxos físicos são:

$$
F_\rho = m,
$$

$$
F_m = \frac{m^2}{\rho}+p,
$$

$$
F_E = (E+p)\frac{m}{\rho}.
$$

Da referência [5], usamos três elementos principais do modelo HEM: a definição da energia total, o cálculo da pressão por equação de estado real e a inclusão do atrito de parede. A energia total é escrita como:

$$
E = \rho\left(e+\frac{1}{2}u^2\right),
$$

onde $e$ é a energia interna específica. Portanto, a partir de $\rho$, $m$ e $E$, calculamos:

$$
u=\frac{m}{\rho},
$$

$$
e=\frac{E}{\rho}-\frac{1}{2}u^2.
$$

A pressão é obtida por uma equação de estado real. Em [5], usa-se principalmente GERG-2008; em um caso próximo ao ponto crítico, usa-se Span-Wagner. Na nossa implementação, usamos CoolProp como aproximação prática inicial:

$$
p = p_{\mathrm{EOS}}(\rho,e).
$$

O modelo de [5] também inclui atrito de parede e, para o escoamento bifásico, usa a correlação de Friedel. Como, nesta etapa, ainda não estamos resolvendo explicitamente as frações de fase necessárias para aplicar Friedel de forma completa, usamos temporariamente uma aproximação do tipo Darcy-Weisbach:

$$
\mathcal{F}
=
\frac{f_D}{2D}\rho u|u|.
$$

## Euler explícito no tempo

Para uma variável conservada genérica $U$, usamos:

$$
\frac{\partial U}{\partial t}
\approx
\frac{U_i^{n+1}-U_i^n}{\Delta t}.
$$

## Diferença centrada no espaço

Para um fluxo genérico $F(U)$, usamos inicialmente:

$$
\frac{\partial F}{\partial x}
\approx
\frac{F_{i+1}^n-F_{i-1}^n}{2\Delta x}.
$$

Assim, a atualização geral fica:

$$
U_i^{n+1}
=
U_i^n
-
\frac{\Delta t}{2\Delta x}
\left(F_{i+1}^n-F_{i-1}^n\right)
+
\Delta t\,S_i^n.
$$

## Primeira equação: massa

A equação é:

$$
\frac{\partial \rho}{\partial t}
+
\frac{\partial m}{\partial x}
=0.
$$

Aplicando Euler explícito e diferença centrada:

$$
\frac{\rho_i^{n+1}-\rho_i^n}{\Delta t}
+
\frac{m_{i+1}^n-m_{i-1}^n}{2\Delta x}
=0.
$$

Portanto:

$$
\rho_i^{n+1}
=
\rho_i^n
-
\frac{\Delta t}{2\Delta x}
\left(m_{i+1}^n-m_{i-1}^n\right).
$$

## Segunda equação: quantidade de movimento

A equação é:

$$
\frac{\partial m}{\partial t}
+
\frac{\partial}{\partial x}
\left(\frac{m^2}{\rho}+p\right)
=-\mathcal{F}.
$$

Aplicando Euler explícito e diferença centrada:

$$
\frac{m_i^{n+1}-m_i^n}{\Delta t}
+
\frac{1}{2\Delta x}
\left[
\left(\frac{m^2}{\rho}+p\right)_{i+1}^n
-
\left(\frac{m^2}{\rho}+p\right)_{i-1}^n
\right]
=-\mathcal{F}_i^n.
$$

Isolando $m_i^{n+1}$:

$$
m_i^{n+1}
=
m_i^n
-
\frac{\Delta t}{2\Delta x}
\left[
\left(\frac{(m_{i+1}^n)^2}{\rho_{i+1}^n}+p_{i+1}^n\right)
-
\left(\frac{(m_{i-1}^n)^2}{\rho_{i-1}^n}+p_{i-1}^n\right)
\right]
-
\Delta t\,\mathcal{F}_i^n.
$$

## Terceira equação: energia

A equação é:

$$
\frac{\partial E}{\partial t}
+
\frac{\partial}{\partial x}
\left((E+p)\frac{m}{\rho}\right)
=0.
$$

Aplicando Euler explícito e diferença centrada:

$$
\frac{E_i^{n+1}-E_i^n}{\Delta t}
+
\frac{1}{2\Delta x}
\left[
\left((E+p)\frac{m}{\rho}\right)_{i+1}^n
-
\left((E+p)\frac{m}{\rho}\right)_{i-1}^n
\right]
=0.
$$

Portanto:

$$
E_i^{n+1}
=
E_i^n
-
\frac{\Delta t}{2\Delta x}
\left[
\left((E_{i+1}^n+p_{i+1}^n)\frac{m_{i+1}^n}{\rho_{i+1}^n}\right)
-
\left((E_{i-1}^n+p_{i-1}^n)\frac{m_{i-1}^n}{\rho_{i-1}^n}\right)
\right].
$$

Assim, a implementação passa a seguir a estrutura conservativa do HEM apresentada em [5], com a ressalva de que a EOS GERG-2008/Span-Wagner é aproximada por CoolProp e o atrito de Friedel é substituído temporariamente por uma forma Darcy-Weisbach.


# Resultados

## Diferença centrada x upwind: caso escalar com py-pde

Antes de resolver o sistema completo de Euler 1D, começamos com um caso simples de uma única equação: a advecção da massa específica com velocidade positiva. A equação considerada é:

$$
\frac{\partial \rho}{\partial t}
+u\frac{\partial \rho}{\partial x}=0.
$$

Neste teste, usamos o `py-pde` para avançar a solução no tempo com Euler explícito. A diferença centrada foi implementada com `state.gradient()`, enquanto o upwind foi implementado manualmente dentro de `evolution_rate()`, usando os valores do campo armazenados em `state.data`.

Para comparar os resultados, usamos também a solução analítica do problema de advecção linear. Essa solução vem do método das características, como apresentado por LeVeque para a equação de advecção na Seção 2.1 de [3]. A mesma ideia de deslocar exatamente o perfil inicial também aparece no esquema de advecção do método PPM de Colella e Woodward [6]. Para a equação

$$
\frac{\partial \rho}{\partial t}
+
u_0\frac{\partial \rho}{\partial x}=0,
$$

as características satisfazem:

$$
\frac{dx}{dt}=u_0.
$$

Ao longo dessas retas, a massa específica permanece constante. Portanto, a solução é simplesmente o perfil inicial deslocado para a direita:

$$
\rho(x,t)=\rho(x-u_0t,0).
$$

No teste, o domínio começa preenchido com $\rho_{alto}$, enquanto a condição de contorno à esquerda impõe $\rho_{baixo}$. Como $u_0>0$, a informação entra pela esquerda e a frente analítica se desloca como:

$$
x_f(t)=u_0t.
$$

Assim, a solução analítica usada na comparação é:

$$
\rho(x,t)=
\begin{cases}
\rho_{baixo}, & x\leq u_0t,\\
\rho_{alto}, & x>u_0t.
\end{cases}
$$

Os dois casos foram rodados com os mesmos parâmetros físicos do Teste 8 e com número CFL igual a 0.90. A linha preta tracejada representa a solução analítica.

![Comparação py-pde entre solução analítica, diferença centrada e upwind](../resources/figures/passo_2_pypde_comparacao_perfis.png)

A diferença centrada usa informações simétricas dos pontos vizinhos. Embora seja uma aproximação de segunda ordem no espaço, ela não respeita a direção preferencial de propagação da informação. Por isso, no caso advectivo, aparecem oscilações numéricas não físicas. Esse comportamento aparece também nos erros finais, que crescem muito para o esquema centrado.

No instante final, os erros foram:

- Diferença centrada: $L^1 = 2.8894\times 10^{14}$ e $L^\infty = 2.3507\times 10^{15}$;
- Upwind: $L^1 = 7.3615$ e $L^\infty = 268.97$.

O método upwind, por outro lado, usa a direção do escoamento para escolher a diferença espacial. Como a velocidade é positiva, a informação vem da esquerda, e a derivada é aproximada por:

$$
\frac{\partial \rho}{\partial x}
\approx
\frac{\rho_i^n-\rho_{i-1}^n}{\Delta x}.
$$

O resultado fica mais difusivo, mas muito mais estável. Esse comportamento justifica o uso de esquemas upwind ou fluxos numéricos direcionais em problemas hiperbólicos, especialmente quando há frentes abruptas, ondas ou descontinuidades.


## Sistema completo

Depois do teste escalar, avançamos as três variáveis conservadas em conjunto:

$$
U =
\begin{bmatrix}
\rho\\
m\\
E
\end{bmatrix},
\qquad
F(U)=
\begin{bmatrix}
m\\
\dfrac{m^2}{\rho}+p\\
(E+p)\dfrac{m}{\rho}
\end{bmatrix},
$$

Nesta etapa, a pressão continua sendo calculada com CoolProp a partir de $\rho$ e $e$, com

$$
u=\frac{m}{\rho},
\qquad
e=\frac{E}{\rho}-\frac{1}{2}u^2.
$$

Para avançar o sistema completo, usamos o fluxo de Rusanov, também chamado de local Lax-Friedrichs:

$$
F_{i+1/2}
=
\frac{1}{2}(F_L+F_R)
-
\frac{1}{2}a_{\max}(U_R-U_L),
$$

com

$$
a_{\max}
=
\max(|u_L|+c_L,\ |u_R|+c_R).
$$

Esse fluxo é mais difusivo do que o HLLC usado em [5], mas é mais simples e robusto para uma primeira implementação do sistema completo. A condição à direita foi tratada como parede fechada por célula fantasma refletindo a quantidade de movimento. Na extremidade aberta à esquerda, ainda usamos uma célula fantasma simplificada de baixa pressão; portanto, esta etapa ainda não reproduz a condição de contorno de pressão de choking usada no artigo.

![Perfis espaciais com Rusanov](../resources/figures/passo_4_perfis_espaciais_rusanov.png)

Também extraímos as variáveis no ponto $x=0{,}08$ m, que corresponde à posição usada no artigo para comparar a pressão perto da extremidade aberta. A simulação foi rodada de $0$ a $30$ ms com malha refinada, usando $N=500$, e os valores foram interpolados no ponto do sensor.

![Variáveis em x=0,08 m](../resources/figures/passo_4_variaveis_008m_refinado.png)

As curvas obtidas apresentam queda monotônica de pressão, massa específica e energia. Esse comportamento difere da curva experimental do Teste 8 em [5], onde aparece uma queda inicial de pressão seguida de recuperação. Essa diferença é esperada nesta etapa, pois ainda não incluímos o modelo de mudança de fase fora do equilíbrio, o termo de transferência de massa do HRM*, a condição de contorno Bernoulli-choking-pressure, nem os modelos completos de atrito e transferência de calor usados em [5].


# Conclusão

Neste trabalho, partimos das equações de Euler unidimensionais com termos fonte e organizamos a formulação em termos das variáveis conservadas $\rho$, $m=\rho u$ e $E$. A partir dessa forma conservativa, discutimos a interpretação física das equações, a relação com volumes de controle e a importância de escrever os fluxos de maneira compatível com problemas hiperbólicos.

Em seguida, estudamos o comportamento numérico de esquemas simples para uma equação de advecção. A comparação entre diferença centrada e diferença upwind mostrou que a diferença centrada, embora tenha maior ordem formal no espaço, pode produzir oscilações não físicas em problemas advectivos. O esquema upwind, por outro lado, introduz difusão numérica, mas respeita a direção de propagação da informação e apresentou comportamento mais estável para o problema considerado.

Depois desse teste escalar, avançamos para a solução conjunta das três equações conservativas. Nessa etapa, a pressão foi obtida a partir da equação de estado via CoolProp e o sistema foi resolvido com um fluxo de Rusanov. Os resultados em perfis espaciais e no ponto $x=0{,}08$ m mostraram uma evolução consistente das variáveis $\rho$, $E$ e $p$, mas ainda com diferenças importantes em relação aos resultados experimentais e numéricos de [5].

Como trabalhos futuros, o próximo passo é incorporar os elementos físicos e numéricos ainda ausentes em relação a [5], incluindo o modelo de mudança de fase fora do equilíbrio, o termo de transferência de massa do HRM*, a condição de contorno Bernoulli-choking-pressure e os modelos de atrito e transferência de calor. Com essas etapas implementadas, pretende-se reproduzir os experimentos do artigo usando o método de volumes finitos e comparar quantitativamente os resultados obtidos com os dados de referência.
